```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3 style="color: #FFC300">VII. Model Evaluation</h3>

```python
# Load the data
model_results = pd.read_csv('../data/processed/metric_scores.csv')
model_results
```

```python
model_results['Accuracy'] = model_results['Accuracy'].apply(lambda x: round(x, 3))
model_results['Precision'] = model_results['Precision'].apply(lambda x: round(x, 3))
model_results['Recall'] = model_results['Recall'].apply(lambda x: round(x, 3))
model_results['F1 Score'] = model_results['F1 Score'].apply(lambda x: round(x, 3))
model_results
```

```python
model_results_melted = model_results.melt(id_vars='Model', var_name='Metric', value_name='Value')

plt.figure(figsize=(12, 8))
sns.barplot(x='Model', y='Value', hue='Metric', data=model_results_melted)
plt.title('Model Performance')
plt.xlabel('Model')
plt.ylabel('Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Metric', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
